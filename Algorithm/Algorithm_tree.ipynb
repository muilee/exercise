```python
# BFS iterative
import Queue

def breathSearch(root):
    if root is None:
        return None
    queue = Queue()
    queue.put(root)
    li = []
    while queue.empty() is not True:
        node = queue.get()
        li.append(node.val)
        if node.left is not None:
            queue.put(node.left)
        if node.right is not None:
            queue.put(node.right)
    return li
```
